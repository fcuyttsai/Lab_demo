```python
import os,sys, time
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,callbacks, models
from functools import partial
import matplotlib.pyplot as plt

import numpy as np
sys.path.append('code')
from tfrecord import *  
from tensorflow.keras.models import load_model
datapath='database' #Data path

os.listdir(datapath) #print out the files in the target folder

model_ = load_model('CNN_model.h5')
train_dataset = get_dataset(datapath+'/train_data.tfrecord',batchsize=128)
```

```python
train_iterator= iter(train_dataset)
for sample in train_iterator:
    test_inputs=sample[0]
    test_outputs=sample[1]
    print(np.shape(test_inputs))
    print(np.shape(test_outputs))
    predictions = model_.predict(test_inputs)
    predicted_labels = np.argmax(predictions, axis=1)
    loss, accuracy = model_.evaluate(test_inputs, test_outputs)
    print("Check Loss:", loss)
    print("Check Accuracy:", accuracy)
    Myplots_V2(test_inputs,predicted_labels,test_outputs, 'Train data')
```
